```python
%matplotlib inline
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-1,1,1000)
x = t**2+4
y = t**3
```

```python
plt.plot(x,y)
```

```python
X = np.array([x,y])
```

```python

A = np.array([[1,-3],[1,4]])
```

```python
egg = A@X
```
